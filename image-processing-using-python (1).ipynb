**Image channels**

THere are 3 channels that represent red blue green , any combination of those three channels can actually represent all the colours that your eye can perceive

```python
#Display RGB channels of our image
#subplot(1,3) means creating 1X3 grid subplot
fig,axs=plt.subplots(1,3,figsize=(15,5))
#to show the image of this matplotlib only the first channel and the colour we use is red to visualize this
axs[0].imshow(img_mp1[:,:,0],cmap='Reds')
axs[1].imshow(img_mp1[:,:,1],cmap='Greens')
axs[2].imshow(img_mp1[:,:,2],cmap='Blues')
#to turn off the grid values
axs[0].axis('off')
axs[1].axis('off')
axs[2].axis('off')
axs[0].set_title('Red channel')
axs[1].set_title('Green channel')
axs[2].set_title('Blue channel')
plt.show()
```

**Matplotlib vs cv2 Numpy Arrays**

* cv2 reads in channels as BGR
* matplotlib reads in channels as RGB

```python
fig,axs=plt.subplots(1,2,figsize=(10,5))
axs[0].imshow(img_cv2)
axs[1].imshow(img_mp1)
axs[0].axis('off')
axs[1].axis('off')
axs[0].set_title('CV Image')
axs[1].set_title('Matplotlib Image')
plt.show()
```

```python
#converting from BGR to RGB 
#by using cv2 convert color, we provide cv2 image and then we use cv2's color bgr to rgb
img_cv2_rgb=cv2.cvtColor(img_cv2,cv2.COLOR_BGR2RGB)
fig,axs=plt.subplots()
axs.imshow(img_cv2_rgb)
axs.axis('off')
plt.show()
```

**Resizing and scaling of an image**

```python
#cv2.resize(image,new size,)or we can reduce the size by a certain percentage using fx and fy so 
#we're going to convert this image into quarter of the size that it was before and then we are going to 
#use our same plot
img_resized=cv2.resize(img,None,fx=0.25,fy=0.25)
fig,axs=plt.subplots(figsize=(8,8))
axs.imshow(img_resized)
axs.axis('off')
plt.show()
```

```python
img_resized.shape
```

```python
#using the same dimensions to stretch out the dimensions here instead of giving fx and fy
#we directly give height and width dimensions
img_resized=cv2.resize(img,(100,200))
fig,axs=plt.subplots(figsize=(8,8))
axs.imshow(img_resized)
axs.axis('off')
plt.show()
```

```python
# In machine learning you want to upscale or resize the image to a specific size that the model expects
#as an input so this is one of the ways you can do it
```

```python
#For upscaling you need to pick an interpolation type that the mod that cv2 will use in order to know 
#how to stretch these pixels out from their original size 
img_resized=cv2.resize(img,(5000,5000),interpolation= cv2.INTER_CUBIC)
fig,axs=plt.subplots(figsize=(8,8))
axs.imshow(img_resized)
axs.axis('off')
plt.show()
```

**Sharpening and Blurring**

```python
#by applying different kernels(matrix) you get different results
#Sharpen Image
kernel_sharpening=np.array([[-1,-1,-1],[-1,9,-1],[-1,-1,-1]])
sharpened_img=cv2.filter2D(img,-1,kernel_sharpening)
#-1 indicating the depth
fig,axs=plt.subplots(figsize=(8,8))
axs.imshow(sharpened_img)
axs.axis('off')
axs.set_title('Sharpend Image')
plt.show()
```

```python
#Blurring image
kernel_3x3=np.ones((3,3),np.float32)/9
blurred_img=cv2.filter2D(img,-1,kernel_3x3)
fig,axs=plt.subplots(figsize=(8,8))
axs.imshow(blurred_img)
axs.axis('off')
axs.set_title('Blurred image')
plt.show()
```

```python
# we can make the image more blur by increasing the number / more than 9 less blur by decreasing
kernel_3x3=np.ones((3,3),np.float32)/100
blurred_img=cv2.filter2D(img,-1,kernel_3x3)
fig,axs=plt.subplots(figsize=(8,8))
axs.imshow(blurred_img)
axs.axis('off')
axs.set_title('Blurred image')
plt.show()
```

```python
kernel_3x3=np.ones((3,3),np.float32)/5
blurred_img=cv2.filter2D(img,-1,kernel_3x3)
fig,axs=plt.subplots(figsize=(8,8))
axs.imshow(blurred_img)
axs.axis('off')
axs.set_title('Blurred image')
plt.show()
```
